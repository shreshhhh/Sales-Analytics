```python
## Data Cleaning
```

```python
sales = df.copy()
```

```python
sales.shape
```

```python
sales.isnull().sum()
```

```python
sales["Order Date"] = pd.to_datetime(sales["Order Date"] ,format = "mixed" )
sales["Ship Date"]  = pd.to_datetime(sales["Ship Date"] ,format = "mixed" ) 
```

```python
sales[["Order Date" , "Ship Date"]].dtypes
```

```python
# Feature Engineering 
```

```python
sales["Year"] = sales["Order Date"].dt.year
```

```python
sales["Quarter"] = sales["Order Date"].dt.quarter
```

```python
sales["Month"] = sales["Order Date"].dt.month
```

```python
sales["Month Name"] = sales["Order Date"].dt.month_name()
```

```python
sales["Day Name"] = sales["Order Date"].dt.day_name()
```

```python
# Shipping Days

sales["Shipping Days"] = (
    sales["Ship Date"] - sales["Order Date"]
).dt.days
```

```python
# Profit Margin

sales["Profit Margin%"] = ( 
    sales["Profit"] / sales["Sales"] * 100 
).round(2)
```
