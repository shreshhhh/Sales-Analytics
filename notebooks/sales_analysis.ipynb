```python
## Data Cleaning
```

```python
sales = df.copy()
```

```python
sales.shape
```

```python
sales.isnull().sum()
```

```python
sales["Order Date"] = pd.to_datetime(sales["Order Date"] ,format = "mixed" )
sales["Ship Date"]  = pd.to_datetime(sales["Ship Date"] ,format = "mixed" ) 
```

```python
sales[["Order Date" , "Ship Date"]].dtypes
```

```python
# Feature Engineering 
```

```python
sales["Year"] = sales["Order Date"].dt.year
```

```python
sales["Quarter"] = sales["Order Date"].dt.quarter
```

```python
sales["Month"] = sales["Order Date"].dt.month
```

```python
sales["Month Name"] = sales["Order Date"].dt.month_name()
```

```python
sales["Day Name"] = sales["Order Date"].dt.day_name()
```

```python
# Shipping Days

sales["Shipping Days"] = (
    sales["Ship Date"] - sales["Order Date"]
).dt.days
```

```python
# Profit Margin

sales["Profit Margin%"] = ( 
    sales["Profit"] / sales["Sales"] * 100 
).round(2)
```

```python
# Exploratory Data Analysis ||EDA
```

```python
sales = pd.read_csv("../data/cleaned_superstore.csv")
sales["Order Date"] = pd.to_datetime(sales["Order Date"])
sales["Ship Date"] = pd.to_datetime(sales["Ship Date"])  
```

```python
# Overall Business Performance
```

```python
total_sales = sales["Sales"].sum()
total_profit = sales["Profit"].sum()
total_orders = sales["Order ID"].nunique()
total_customers = sales["Customer ID"].nunique()
avg_order_value = total_sales / total_orders
profit_margin = (total_profit / total_sales) * 100

print(f"Total Sales           : ${total_sales : ,.2f}")
print(f"Total Profit          : ${total_profit  : ,.2f}")
print(f"Total Orders          : ${total_orders }")
print(f"Total Customers       : ${total_customers : ,.2f}")
print(f"Average Order Value   : ${avg_order_value : ,.2f}")
print(f"Profit Margin         : ${profit_margin : ,.2f}")
```

```python
monthly_sales = (
    sales.groupby("Month")["Sales"]
    .sum()
    .reset_index()
)

monthly_sales
```

```python
plt.figure(figsize=(10,5))

plt.plot(
    monthly_sales["Month" ],
    monthly_sales["Sales"],
    marker = 'o'
)

plt.title("Montly Sales")
plt.xlabel("Month")
plt.ylabel("Sales")

plt.show()
```

```python
# Sales by Category 

category_sales = (
    sales.groupby("Category")[["Sales" , "Profit"]]
    .sum()
    .sort_values("Sales" , ascending= False)
)

category_sales
```

```python
category_sales["Sales"].plot(
    kind="bar" , 
    figsize = ( 8 , 5 )
)

plt.title("Sales by Category")
plt.ylabel("Sales")
plt.xticks(rotation = 0 )

plt.show()
```

```python
region_sales = (
    sales.groupby("Region")[["Sales" , "Profit"]] 
    .sum()
    .sort_values("Sales" , ascending=False)
)
region_sales
```

```python
region_sales["Sales"].plot(
    kind= "bar" , 
    figsize=(8,5)
)
plt.title("Regional Sales")
plt.ylabel("Sales")
plt.xticks(rotation = 0)

plt.show()
```

```python
top_customers = (
    sales.groupby("Customer Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)
top_customers
```

```python
top_customers.sort_values().plot(
    kind="barh" , 
    figsize= (8,6)
)

plt.title("Top 10 Customers")
plt.xlabel("Sales")

plt.show()
```

```python
top_products = (
    sales.groupby("Product Name")["Sales"]
    .sum()
    .sort_values(ascending=False)
    .head(10)
)

top_products
```

```python
top_products.sort_values().plot(
    kind="barh",
    figsize=(9,6)
)

plt.title("Top selling Products")
plt.xlabel("Sales")

plt.show()
```

```python
## Discount vs Profit

plt.figure(figsize=(8,5))

plt.scatter(
    sales["Discount"] ,
    sales["Profit"] ,
    alpha= 0.5
)

plt.xlabel("Discount")
plt.ylabel("Profit")
plt.title( "Discount vs Profit" )

plt.show()
```

```python
numeric_cols = sales.select_dtypes(include=np.number)

corr = numeric_cols.corr()

corr
```

```python
plt.Figure(figsize=(10,6))

sns.heatmap(
    corr , 
    annot=True ,
    cmap = "coolwarm" , 
    fmt = ".2f" , 
)

plt.title("Correlation Matrix")

plt.show()
```
